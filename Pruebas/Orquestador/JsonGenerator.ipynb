```python
import random
import json
from pathlib import Path
from sklearn.model_selection import train_test_split

random.seed(42)

def build_dataset_from_intents(intents):
    dataset = []
    for intent in intents:
        for question in intent["examples"]:
            entry = {
                "input": question,
                "response": json.dumps({
                    "tools": intent["tools"],
                    "justification": intent["justification"]
                }, ensure_ascii=False)
            }
            dataset.append(entry)
    return dataset

# -----------------------------
# División estratificada
# -----------------------------
def stratified_split(dataset, intents, train_ratio=0.8):
    # Agrupar por clase (tools)
    grouped = {}
    for intent in intents:
        key = tuple(intent["tools"])
        grouped[key] = []

    for item in dataset:
        tools_key = tuple(json.loads(item["response"])["tools"])
        grouped[tools_key].append(item)

    train_data, val_data = [], []
    for group in grouped.values():
        train, val = train_test_split(group, train_size=train_ratio, random_state=42)
        train_data.extend(train)
        val_data.extend(val)

    # Shuffle final
    random.shuffle(train_data)
    random.shuffle(val_data)
    return train_data, val_data

# -----------------------------
# Guardado a .jsonl
# -----------------------------
def save_jsonl(data, path):
    with open(path, "w", encoding="utf-8") as f:
        for item in data:
            f.write(json.dumps(item, ensure_ascii=False) + "\n")

# -----------------------------
# Ejecución principal
# -----------------------------
dataset = build_dataset_from_intents(intents)
# train_set, val_set = stratified_split(dataset, intents)

# output_dir = Path("fine_tuning_data")
# output_dir.mkdir(exist_ok=True)

# save_jsonl(train_set, output_dir / "train.jsonl")
# save_jsonl(val_set, output_dir / "validation.jsonl")

# print(f"✅ Generados {len(train_set)} ejemplos de entrenamiento y {len(val_set)} de validación.")
```

```python
# !pip install transformers==4.41.2 accelerate sentencepiece --quiet
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# tokenizer = AutoTokenizer.from_pretrained("seduerr/mt5-paraphrases-espanol")
paraphraser = AutoModelForSeq2SeqLM.from_pretrained("seduerr/mt5-paraphrases-espanol")
```

```python

def paraphrase_batch(sentences):
    """Genera parafraseos en batch para ahorrar tiempo"""
    prompts = [f"Parafrasea: {s}" for s in sentences]
    outputs = paraphraser(
        prompts,
    )

    # outputs es una lista de dicts, hay que agrupar por input
    results = []
    for i in range(0, len(outputs)):
        variants = [o["generated_text"] for o in outputs[i:i+1]]
        results.append(variants)
    return results

# Ejemplo de uso
dataset_prueba = [
    "¿Qué pone en el cartel?",
    "¿Puedes leer el letrero?",
    "¿Qué dice el texto que aparece en la imagen?"
]

paraphrased = paraphrase_batch(dataset_prueba)

for original, variants in zip(dataset_prueba, paraphrased):
    print(f"Original: {original}")
    for v in variants:
        print(f" - {v}")
```

```python

def paraphrase_examples(example, num_variants=3):
    augmented = []
    prompt = f"Parafrasea esta petición de un cliente de forma natural y breve y devuelve únicamente la nueva frase: '{example['input']}'"
    outputs = paraphraser(prompt, max_new_tokens=64, num_return_sequences=num_variants, do_sample=True, top_k=50)
    for o in outputs:
        text = o["generated_text"].strip()
        if text and text != example['input']:
            augmented.append({"input":text,
                            "response": example['response']})
    return augmented

for example in dataset:
    augmented = paraphrase_examples(example, num_variants=2)
    dataset.extend(augmented)

```
